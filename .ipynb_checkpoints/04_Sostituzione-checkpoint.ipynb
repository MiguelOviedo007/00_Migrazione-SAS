```python
import os
import fileinput
import re

def sostituisci_stringhe(file_path, sostituzioni):
    with fileinput.FileInput(file_path, inplace=True, backup='', mode='r') as file:
        for line in file:
            for vecchia, nuova in sostituzioni.items():
                pattern = re.compile(re.escape(vecchia), re.IGNORECASE)
                line = pattern.sub(nuova, line)
            print(line, end='')

def leggi_e_modifica_file(cartella, estensione, sostituzioni):
    for root, dirs, files in os.walk(cartella):
        for file_name in files:
            if file_name.endswith(estensione):
                file_path = os.path.join(root, file_name)
                sostituisci_stringhe(file_path, sostituzioni)
                print(f"Sostituzione effettuata in {file_path}")
```
