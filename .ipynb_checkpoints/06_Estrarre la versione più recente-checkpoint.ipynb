```python
import os
import re

def clean_sas_files(folder_path):
    for root, dirs, files in os.walk(folder_path):
        sas_files = [file for file in files if file.endswith(".sas")]

        # Raggruppa i file per nome base
        grouped_files = {}
        for sas_file in sas_files:
            match = re.match(r'^(.+)_v(\d+\.\d+).sas$', sas_file)
            if match:
                base_name, version = match.groups()
                version = float(version)
                if base_name not in grouped_files or grouped_files[base_name][0] < version:
                    grouped_files[base_name] = (version, sas_file)

        # Elimina i file non selezionati
        for base_name, (version, selected_file) in grouped_files.items():
            for sas_file in sas_files:
                if sas_file.startswith(base_name) and sas_file != selected_file:
                    file_path = os.path.join(root, sas_file)
                    os.remove(file_path)
                    print(f"Eliminato: {file_path}")


if __name__ == "__main__":
    folder_path = r"C:\Miguel\progetti\05_Modifica Testi\01_data\TEST\PROVA ELIMINA"
    clean_sas_files(folder_path)
```

```python
import os
import re

def clean_sas_files(folder_path):
    for root, dirs, files in os.walk(folder_path):
        sas_files = [file for file in files if file.endswith(".sas")]

        # Elimina i file che terminano con " - Copia.sas"
        sas_files = [file for file in sas_files if not file.endswith(" - Copia")]

        # Raggruppa i file per nome base
        grouped_files = {}
        for sas_file in sas_files:
            match = re.match(r'^(.+)_v(\d+\.\d+).sas$', sas_file)
            if match:
                base_name, version = match.groups()
                version = float(version)
                if base_name not in grouped_files or grouped_files[base_name][0] < version:
                    grouped_files[base_name] = (version, sas_file)

        # Elimina i file non selezionati
        for base_name, (version, selected_file) in grouped_files.items():
            for sas_file in sas_files:
                if sas_file.startswith(base_name) and sas_file != selected_file:
                    file_path = os.path.join(root, sas_file)
                    os.remove(file_path)
                    print(f"Eliminato: {file_path}")

if __name__ == "__main__":
    folder_path = r"C:\Miguel\progetti\05_Modifica Testi\01_data\TEST\CODICI_SAS"
    clean_sas_files(folder_path)
```

```python
import os
import shutil
import subprocess

def cambia_permessi_cartella(percorso_cartella):
    try:
        subprocess.run(["icacls", percorso_cartella, "/grant", f"{os.getlogin()}:(OI)(CI)F"], check=True)
    except subprocess.CalledProcessError as e:
        print(f"Errore durante la modifica dei permessi della cartella '{percorso_cartella}': {e}")

def elimina_sottocartelle_old(directory):
    contenuto_directory = os.listdir(directory)
    cartelle_da_elimare = ["old", "old_version", "old script", "old dg", "vecchi script", "_old"]

    for elemento in contenuto_directory:
        percorso_elemento = os.path.join(directory, elemento)

        if os.path.isdir(percorso_elemento) and elemento.lower() in cartelle_da_elimare:
            # Cambia i permessi prima di eliminare
            cambia_permessi_cartella(percorso_elemento)

            try:
                # Rimuovi la directory con shutil.rmtree
                shutil.rmtree(percorso_elemento)
                print(f"Sottocartella '{elemento}' rimossa con successo.")
            except Exception as e:
                print(f"Errore durante la rimozione della sottocartella '{elemento}': {e}")

        elif os.path.isdir(percorso_elemento):
            elimina_sottocartelle_old(percorso_elemento)

# Specifica il percorso della directory da esaminare
percorso_directory = r"C:\Miguel\progetti\05_Modifica Testi\01_data\TEST\CODICI_SAS"

# Chiama la funzione per eliminare le sottocartelle specificate
elimina_sottocartelle_old(percorso_directory)
```
